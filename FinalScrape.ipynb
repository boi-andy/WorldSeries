```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Wikipedia page URL with the table you want to scrape
url = "https://en.wikipedia.org/wiki/World_Series"

# Send a GET request to the URL
response = requests.get(url)
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table by its class, id, or other attributes
    # You might need to inspect the HTML source of the page to find the appropriate selector
    table = soup.find('table', {'class': 'wikitable'})

    # Check if the table is found
    if table:
        # Extract data from the table and convert it to a pandas DataFrame
        rows = []
        for row in table.find_all('tr'):
            columns = row.find_all(['th', 'td'])
            row_data = [column.get_text(strip=True) for column in columns]
            rows.append(row_data)

        # Create a DataFrame
        df = pd.DataFrame(rows[1:], columns=rows[0])

        # Print the DataFrame
        print(df)

        # Save the DataFrame as a CSV file
        df.to_csv('output.csv', index=False)
        
        print("Data saved to output.csv.")
    else:
        print("Table not found on the page.")
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```
